```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3)
y = np.sin(x)

plt.xlim(-3, 3)
plt.ylim(-2, 2)

arrowprops1 = {
    "arrowstyle": "->",
    "connectionstyle": "arc3, rad=-0.3",
    "ec": 'black', 
    "lw": 2      
    }

plt.annotate("max: 1.0", xytext=(0, 1.5), xy=(1.5, 1.0),
             arrowprops=arrowprops1, fontsize=20)


arrowprops2 = {
    "arrowstyle": "->",
    "connectionstyle": "arc3, rad=0.3",
    "ec": 'black', 
    "lw": 2      
    }
plt.annotate("min: -1.0", xytext=(-3, -1.5), xy=(-1.5, -1),
             arrowprops=arrowprops2, fontsize=20)


plt.plot(x, y, label='$y=sin(x)$')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

d1 = np.random.normal(-3, 2, 1000)
d2 = np.random.normal(0, 3, 1000)
d3 = np.random.normal(3, .6, 1000)

bins = 50
histtype = 'stepfilled'
alpha = .6

plt.hist(d1, bins=bins, histtype=histtype, 
         alpha=alpha, label='$\mu=-3, \sigma=2$')
plt.hist(d2, bins=bins, histtype=histtype, 
         alpha=alpha, label='$\mu=0, \sigma=3$')
plt.hist(d3, bins=bins, histtype=histtype, 
         alpha=alpha, label='$\mu=3, \sigma=0.6$')

plt.legend()
plt.show()
```
