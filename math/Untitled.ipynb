```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
np.random.seed(123)
```

```python
X = np.random.multivariate_normal([-3, 4], np.eye(2), 20)
Y = np.random.multivariate_normal([3, -2], 2*np.eye(2), 20)
```

```python
plt.rcParams["figure.figsize"] = (10, 8)
```

```python
plt.plot(X, 'r^')
plt.plot(Y, 'g*')
```
